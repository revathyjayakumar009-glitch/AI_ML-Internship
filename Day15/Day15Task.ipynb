**Task 2: Data Practice** 
1. Create list of marks  
2. Find average  
3. Find max and min  
4. Count total students 

```python
marks = [85,65,75,50,99,95,49]
#Average
avg = sum(marks) / len(marks)
print("Average marks:",avg)

#Maximum and Minimum
print("Maximum:",max(marks))
print("Minimum:",min(marks))

#Count total students
print("Total Students:",len(marks))
```

**Task 3: Data Cleaning**
1. Remove None values  
2. Remove negative values  
3. Clean invalid data 

```python
marks = [60, None, 90, -10, 70, None, 85, -5,95,99,None,-5]
#Remove None values
cleaned = [m for m in marks if m is not None]
print("Cleaned Data:",cleaned)

#Remove Negative
cleaned = [m for m in cleaned if m >=0]
print("Cleaned:",cleaned)
#Clean invalid data
cleaned = [m for m in marks if m is not None and m >= 0]
print("Cleaned data:",cleaned)
```

**Task 4: Data Filtering** 
1. Find passed students  
2. Filter marks > 75  
3. Filter even numbers  

```python
marks = [85,65,75,50,99,95,49]
#Find passed students
passed = [m for m in marks if m >= 50]
print("Passed:", passed)

#Filter marks > 75
greater_75 = [m for m in marks if m>=75]
print("Marks greater than 75:",greater_75)

#Filter even numbers
even_numbers = [m for m in marks if m %2 == 0]
print("Even Numbers:",even_numbers)
```

**Task 5: Data Transformation** 
1. Convert marks to percentage  
2. Multiply values  
3. Normalize data

```python
#Convert marks to percentage
percentage = [m / 100 for m in marks]
print("Percentage:", percentage)

#Multiply values
value_multiply = [m * 2 for m in marks]
print("Multiply values:",value_multiply)

#Normalize data
min_value = min(marks)
max_value = max(marks)

normalized = [(m - min_value) / (max_value - min_value) for m in marks]
print("Normalized:", normalized)
```
