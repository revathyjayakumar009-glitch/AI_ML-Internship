**Task 4: Method Practice**
1. Create method to:  
o Check pass/fail  
o Calculate grade 

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks

    def check_result(self):
        if self.marks >= 40:
            print("Result:","Pass")
        else:
            print("Result:","Fail")

    def calculate_grade(self):
        if self.marks >= 90:
            print("Grade: A")
        elif self.marks >= 75:
            print("Grade: B")
        elif self.marks >= 50:
            print("Grade: C")
        else:
            print("Grade: D")
    
    def display(self):
        print("Student_Details")
        print("Name:",self.name,"\n","Marks:",self.marks)
    
        
name = input("Enter student name: ")
marks = int(input("Enter marks: "))

s1 = Student(name, marks)
s1.display()
s1.check_result()
s1.calculate_grade()
```

**Task 5: Object List** 
1. Create list of students  
2. Print all students  
3. Find topper 

```python
class Student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
        
    def display(self):
        print("Student_Details")
        print("Name:",self.name,"\n","Marks:",self.marks)

Student_list = []  #list of students
n = int(input("Enter number of students: "))

for i in range(n):
     name = input("Enter name: ")
     marks = int(input("Enter marks: "))

     s = Student(name, marks)
     Student_list.append(s)

print("\n--- Student List ---")
for s in Student_list:
    s.display() #print all students

topper = Student_list[0]

for s in Student_list:
    if s.marks > topper.marks:
        topper = s
print("\n--- Topper ---")
print("Name:", topper.name)
print("Marks:", topper.marks)
```

**1. Create class BankAccount**  
o deposit  
o withdraw  
o check balance  
**2. Create class Product** 
o name, price  
o calculate discount

```python
#Bank Account
class BankAccount:
    def __init__(self, name, balance=0):
        self.name = name
        self.balance = balance

    def deposit(self):
        amount = float(input("Enter amount to deposit: "))
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self):
        amount = float(input("Enter amount to withdraw: "))
        if amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def check_balance(self):
        print("Current Balance:", self.balance)

name = input("Enter account holder name: ")
balance = float(input("Enter initial balance: "))
account = BankAccount(name, balance)


while True:
    print("\n--- Bank Menu ---")
    print("1. Deposit")
    print("2. Withdraw")
    print("3. Check Balance")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        account.deposit()
    elif choice == '2':
        account.withdraw()
    elif choice == '3':
        account.check_balance()
    elif choice == '4':
        print("Thank you! Exiting...")
        break
    else:
        print("Invalid choice! Try again.")
```

```python
# Product 
class Product:
    def __init__(self):
        self.name = ""
        self.price = 0
        self.discount_percent = 0
        self.final_price = 0

    def get_data(self):
        self.name = input("Enter product name: ")
        self.price = float(input("Enter product price: "))
        self.discount_percent = float(input("Enter discount %: "))

    def calculate_discount(self):
        discount = (self.price * self.discount_percent) / 100
        self.final_price = self.price - discount

    def display(self):
        print("\n--- Product Details ---")
        print("Product:", self.name)
        print("Original Price:", self.price)
        print("Discount:", self.discount_percent, "%")
        print("Final Price:", self.final_price)

p1 = Product()

p1.get_data()
p1.calculate_discount()
p1.display()
```
