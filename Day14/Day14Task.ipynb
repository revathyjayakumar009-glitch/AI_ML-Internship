**Task 2: JSON Practice** 
1. Convert dictionary to JSON  
2. Convert JSON to dictionary  
3. Write JSON to file  
4. Read JSON from file 

```python
import json
import csv
```

```python
#convert dictionary to json
data_dict = {"Revathy":95,"Aswin":88,"Janaki":99,"Sandra":75}

json_data = json.dumps(data_dict)
print(json_data)
```

```python
#convert json to dictionary

json_data='{"Revathy":95,"Aswin":88,"Janaki":99,"Sandra":75}'
data_dict = json.loads(json_data)
print(data_dict)
```

```python
#writ json to a file
with open("json_data.json","w") as file:
    json.dump(json_data,file)
```

```python
#Read json from file
with open("json_data.json","r") as file:
    data=json.load(file)

print(data)
```

**Task 3: CSV Practice**
1. Create CSV file  
2. Write student data  
3. Read CSV file  
4. Print formatted output  

```python
# Create CSV file
with open("Student.csv","w",newline="") as file:
    writer=csv.writer(file)

#write student data
    writer.writerow(["Name","Marks"])
    writer.writerow(["Revathy",95])
    writer.writerow(["Rahul",88])
    writer.writerow(["Hari",75])
```

```python
#Read CSV file
with open("Student.csv", "r") as file:
    reader = csv.reader(file)

#print formatted output
    for row in reader:
        name = row[0]
        marks = row[1]
        print(f"Name: {name} | Marks: {marks}")
```

**Task 4: Combined Task** 
Program: 
• Read CSV file  
• Convert to dictionary  
• Save as JSON 

```python
#Read CSV and convert it into dictionary
with open("Student.csv", "r") as file:
    reader = csv.DictReader(file)
    data=list(reader)
#save as JSON
with open("students.json", "w") as file:
    json.dump(data, file)

print("CSV data successfully converted to JSON!")
    

```
