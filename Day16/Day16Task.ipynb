**Task 2: Cleaning Practice** 
1. Remove None values  
2. Remove negative values  
3. Keep only valid data 

```python
employees = [
    {"name": "John", "age": 28, "salary": 30000},
    {"name": "Sara", "age": None, "salary": 35000},
    {"name": "Mike", "age": 25, "salary": -20000},
    {"name": "Anna", "age": 30, "salary": 40000}
]
print(employees)
```

```python
#Print valid data
cleaned_employee = [
    emp for emp in employees if emp["age"] is not None and emp["salary"]>=0]
print(cleaned_employee)
                    
                    
```

**Task 3: Transformation**
1. Add pass/fail field  
2. Convert marks to percentage  
3. Create new calculated field  

```python
students = [
    {"name": "A", "marks": 80},
    {"name": "B", "marks": 40},
    {"name": "C", "marks": 65}
]
print(students)
```

```python
#Add pass or fail

for s in students:
    s["result"] = "Pass" if s["marks"] >= 50 else "Fail"

print(students)
```

```python
#convert marks in percentage
total_marks = 100
for s in students:
    s["percentage"] = (s["marks"] / total_marks) * 100

print(students)
```

```python
#create new calculated filed
#Add grade as new calculated
for s in students:
    s["grade"] = "A" if s["marks"] >= 75 else "B" if s["marks"] >= 50 else "C"

print(students)
```

**Task 4: Sorting** 
1. Sort students by marks  
2. Print top 3 students 

```python
students = [
    {"name": "A", "marks": 80},
    {"name": "B", "marks": 40},
    {"name": "C", "marks": 65},
    {"name": "D", "marks": 90},
    {"name": "E", "marks": 55}
]
```

```python
#Sort
students.sort(key=lambda x: x["marks"], reverse=True) 
print(students)
```

```python
#print top 3
top_3=sorted(students,key=lambda x:x["marks"], reverse=True)[:3]
for s in top_3:
    print(s)
```

**Task 5: Grouping** 
1. Group pass and fail  
2. Count each group  

```python
result = {"pass": [], "fail": []} 
 
for s in students: 
    if s["marks"] >= 50: 
        result["pass"].append(s) 
    else: 
        result["fail"].append(s) 
 
print(result)
```

```python
#count each group
print("Pass Count:", len(result["pass"]))
print("Fail Count:", len(result["fail"]))
```

**Task 6: Data Analysis** 
1. Find average marks  
2. Find highest marks  
3. Find lowest marks

```python
#Find average marks
total = sum(s["marks"] for s in students)
average = total / len(students)

print("Average Marks:", average)
```

```python
#Find highest marks
highest = max(s["marks"] for s in students)
print("Highest Marks:", highest)
```

```python
#Find lowest marks
lowest = min(s["marks"] for s in students)
print("Lowest Marks:", lowest)
```
