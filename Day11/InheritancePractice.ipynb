# Inheritance
Inheritance is a feature of Object-Oriented Programming that allows a class (child class) to inherit properties and methods from another class (parent class).


Advantages

1.Code reusability
2.Reducing duplication
3.Better organization

```python
class Person: 
    def __init__(self, name): 
        self.name = name 
 
    def display(self): 
        print("Name:", self.name) 
 
class Student(Person): 
    def __init__(self, name, marks): 
        super().__init__(name) 
        self.marks = marks 
 
s1 = Student("John", 85) 
s1.display() 
print(s1.marks)
```

```python
# 1.Single inheritance
class Person:
    def __init__(self,name):
        self.name = name

class Employee(Person):
    def __init__(self,name,salary):
        super().__init__(name)
        self.salary = salary

#creating objects
# Single inheritance
emp = Employee("john",40000)
print(emp.name,"\t",emp.salary)
```

```python
# 2. Multiple inheritance
class Job:
    def __init__(self,designation):
        self.designation = designation

class EmployeePersonJob(Employee,Job): #Inherits from Employee and job  ---> child
    def __init__(self,name,salary,designation):
        Employee.__init__(self,name,salary)
        Job.__init__(self,designation)

emp = Employee("john",40000)
print(emp.name,"\t",emp.salary)

emp2 = EmployeePersonJob("Appu",60000,"developer")
print(emp2.name,"\t",emp2.salary,"\t",emp2.designation)
```

```python
class Person:
    def __init__(self,name):
        self.name = name

class Employee(Person):
    def __init__(self,name,salary):
        super().__init__(name)
        self.salary = salary

# 2. Multiple inheritance
class Job:
    def __init__(self,designation):
        self.designation = designation

class EmployeePersonJob(Employee,Job): #Inherits from Employee and job  ---> child
    def __init__(self,name,salary,designation):
        Employee.__init__(self,name,salary)
        Job.__init__(self,designation)

# 3. Multilevel inheritance
class Manager(EmployeePersonJob):
    def __init__(self,name,salary,designation,department):
        EmployeePersonJob.__init__(self,name,salary,designation)
        self.department = department

# 4. Hierachical inheritance
class AssistantManager(EmployeePersonJob):
    def __init__(self,name,salary,designation,team_size):
        EmployeePersonJob.__init__(self,name,salary,designation)
        self.team_size = team_size

# 5. Hybrid inheritance(Multiple+Mutilevel)
class SeniorManager(Manager,AssistantManager):
    def __init__(self,name,salary,designation,department,team_size):
        Manager.__init__(self,name,salary,designation,department)
        AssistantManager.__init__(self,name,salary,designation,team_size)
    

    

#creating objects
# 1.Single inheritance
emp1 = Employee("john",40000)
print(emp1.name,"\t",emp1.salary)

# 2. Multiple inheritance
emp2 = EmployeePersonJob("Appu",60000,"developer")
print(emp2.name,"\t",emp2.salary,"\t",emp2.designation)

# 3. Multilevel inheritance
mgr = Manager("Bob",200000,"IT Manager","IT")
print(mgr.name,"\t",mgr.salary,"\t",mgr.designation,"\t",mgr.department)

# 4. Hierachical inheritance
asst_mgr = AssistantManager("Jafer",1500000,"Finance",20)
print(asst_mgr.name,"\t",asst_mgr.salary,"\t",asst_mgr.designation,"\t",asst_mgr.team_size)

# 5. Hybrid inheritance
sen_mgr = SeniorManager("David",250000,"Sr.Manager","HR",25) 
print(sen_mgr.name,"\t",sen_mgr.salary,"\t",sen_mgr.designation,"\t",sen_mgr.department,"\t",sen_mgr.team_size)
```

## Polymorphism
In Python, it allows the same function or method name to behave differently depending on the object or situation.

Types of Polymorphism in Python
️Function Polymorphism

Same function works with different data types

```python
print(len("Hello"))   # String → 5
print(len([1, 2, 3])) # List → 3
```

**Method Overriding (Runtime Polymorphism)**

Child class changes the behavior of a parent class method

```python
class Person: 
    def display(self): 
        print("This is Person") 
 
class Student(Person): 
    def display(self): 
        print("This is Student") 
 
s = Student() 
s.display() 
```

# Encapsulation

Encapsulation means wrapping data (variables) and methods (functions) together into a single unit (class) and restricting direct access to some of the object's components.

Why Encapsulation?

1.Protects data from unauthorized access
2.Improves code security
3.Makes code more organized
4.Provides controlled access using methods

```python
class Student: 
    def __init__(self, name, marks): 
        self.name = name 
        self.__marks = marks   # private variable 
 
    def get_marks(self): 
        return self.__marks 
 
s = Student("Sara", 90) 
print(s.get_marks())
```
