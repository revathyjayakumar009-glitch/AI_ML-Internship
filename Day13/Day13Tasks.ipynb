**Task 3: Optimization Practice** 
1. Replace loops with built-in functions  
2. Optimize sorting  
3. Reduce repeated code 

```python
#Replace loops with built-in functions
#Before
numbers = [1, 2, 3, 4, 5]
total = 0
for num in numbers:
    total += num
print(total)
```

```python
#After
total=sum(numbers)
print(total)
```

```python
#Optimize sorting
#Before
nums = [5, 2, 9, 1]
for i in range(len(nums)):
    for j in range(len(nums)):
        if nums[i] < nums[j]:
            nums[i], nums[j] = nums[j], nums[i]
print(nums)
```

```python
#After
nums.sort()
print(nums)
```

```python
#Reduce repeated code
#Before
print("Total:", 10 + 5)
print("Total:", 20 + 5)
print("Total:", 30 + 5)
```

```python
#After
def calculate_total(value):
    print("Total:", value + 5)

calculate_total(10)
calculate_total(20)
calculate_total(30)
```

**Task 4: Refactoring**
     Convert: 
• Long code → functions  
• Repeated logic → reusable function 

```python
#Long code into functions
#Before
marks1 = [80, 75, 90]
total1 = 0
for m in marks1:
    total1 += m
avg1 = total1 / len(marks1)
print("Student 1 Average:", round(avg1,2))

#before
marks2 = [60, 70, 65]
total2 = 0
for m in marks2:
    total2 += m
avg2 = total2 / len(marks2)
print("Student 2 Average:", avg2)
```

```python
#After
def calculate_average(marks):
    return sum(marks) / len(marks)

def display_average(student_name, marks):
    avg = calculate_average(marks)
    print(f"{student_name} Average:", round(avg,2))

display_average("Student 1", [80, 75, 90])
display_average("Student 2", [60, 70, 65])
```

```python
#Repeated logic--> Reusable function
#Before
marks = 85
if marks >= 50:
    print("Pass")
else:
    print("Fail")

marks = 40
if marks >= 50:
    print("Pass")
else:
    print("Fail")

marks = 70
if marks >= 50:
    print("Pass")
else:
    print("Fail")
```

```python
#After
def check_result(marks):
    if marks >= 50:
        print("Pass")
    else:
        print("Fail")

check_result(85)
check_result(40)
check_result(70)
```
