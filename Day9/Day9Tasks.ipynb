**Task 2: File Data Practice** 
1. Create file with student data  
2. Read and store in list of dictionary  
3. Print all students  
4. Find average marks  
5. Find topper 

```python
#Create file with student data
n=int(input("Enter the number of Students:"))

with open("Student_Mark.txt","w") as file:
    for i in range(n):
        name = input("Enter the student name:")
        mark = int(input("Enter the marks:"))

        file.write(f"{name},{mark}\n")

print("File is created with student data")
    
                   
        
    
      
```

```python
#Read and Store in List of Dictionary
Student=[]
with open("Student_Mark.txt","r") as file:
    for line in file: 
        name, marks = line.strip().split(",") 
        Student.append({"name": name, "marks": int(marks)}) 
 
print(Student)
```

```python
#Print all students

with open("Student_Mark.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
#Find Average mark
total = sum(s["marks"] for s in Student) 
avg = total / len(Student) 
print("Average:", avg) 
```

```python
#Find topper
top = max(Student, key=lambda x: x["marks"]) 
print("Topper:", top["name"])
```

**Task 3: Data Processing Programs**
1. Filter students with marks > 75  
2. Count number of students  
3. Find lowest marks  
4. Sort students by marks 

```python
Student=[]
with open("Student_Mark.txt","r") as file:
    for line in file: 
        name, marks = line.strip().split(",") 
        Student.append({"name": name, "marks": int(marks)}) 
 
print(Student)
```

```python
#Filter Student with marks > 75
print("Students with marks > 75")
for student in Student:
    if student["marks"] > 75:
        print(student["name"],student["marks"])
```

```python
#Count number of students
print("Total number of students:",len(Student))
```

```python
#FInd the lowest mark
Min_Of_Marks = min(Student, key=lambda x:x["marks"])

print("Lowest Marks")
print(Min_Of_Marks["name"],Min_Of_Marks["marks"])
```

```python
#Sort students by marks
sorted_students = sorted(Student, key=lambda x:x["marks"])

print("Students sorted by marks")
for student in sorted_students:
    print(student["name"],student["marks"])
```

```python
#function to filter student
def filter_students(Student, limit):
    filtered = []
    
    for s in Student:
        if s["marks"] > limit:
            filtered.append(s)
    
    return filtered
result = filter_students(Student, 75)

print("Students with marks > 75:")
for s in result:
    print(s["name"], s["marks"])
    
```

**Task 5: Logic Challenge** 
1. Find student with second highest marks  
2. Count number of passed/failed students  
3. Convert data into dictionary format 

```python
#Find student with second highest marks
def second_highest(Student):
    unique_mark = list(set(s["marks"] for s in Student))
    unique_mark.sort(reverse=True)

    s_highest = unique_mark[1]

    for s in Student:
        if s["marks"] == s_highest:
            return s


print("Second highest")

s = second_highest(Student)
print("Second Topper:", s["name"], s["marks"])
```

```python
#Count number of passed and failed students
def count_pass_fail(Student):
    passed = 0
    failed = 0
    
    for s in Student:
        if s["marks"] >= 60:
            passed += 1
        else:
            failed += 1
    
    return passed, failed

p, f = count_pass_fail(Student)
print("Passed:", p)
print("Failed:", f)
```

```python
#Convert data into dictionary format
def convert_to_dict(Student):
    return {s["name"]: s["marks"] for s in Student}
student_dict = convert_to_dict(Student)
print(student_dict)
```
