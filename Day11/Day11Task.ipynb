**Task 2: Inheritance Practice** 
1. Create class Animal  
2. Create class Dog inheriting Animal  
3. Add method sound() 

```python
class Animal:
    def sound(self):
        print("Animal make sounds")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

**Task 3: Method Overriding**
1. Parent class → Vehicle  
2. Child class → Car  
3. Override method start() 

```python
class Vehicle:
    def start(self):
        print("Vehicle is Ready to satrt")

class Car(Vehicle):
    def start(self):
        print("Car starts")

c=Car()
c.start()
```

**Task 4: Encapsulation Practice** 
1. Create class BankAccount  
2. Make balance private  
3. Create methods:  
o deposit  
o withdraw  
o check balance  

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def get_input(self):
        self.__balance = int(input("Enter the initial amount: "))

    def deposit(self):
        amount = int(input("Enter the amount to be deposited: "))
        if amount > 0:
            self.__balance += amount
            print("Deposited:", amount)
        else:
            print("Invalid amount")

    def withdraw(self):
        amount = int(input("Enter the amount to be withdrawn: "))
        if amount <= self.__balance:
            self.__balance -= amount   # ✅ FIXED
            print("Withdrawn:", amount)
        else:
            print("Insufficient Balance")

    def display(self):
        print("Current Balance:", self.__balance)


acc = BankAccount()

acc.get_input()
acc.deposit()
acc.withdraw()
acc.display()
```

**Task 5: Combined Practice** 
Create: 
• Class Person  
• Class Employee (inherits Person)  
• Add:  
o name  
o salary  
o display details

```python
class Person:
    def __init__(self):
        self.name = ""

    def get_input_data(self):
        self.name = input("Enter name: ")


class Employee(Person):
    def __init__(self):
        super().__init__()# super() is used to access methods and constructors of the parent class from the child class.
        self.salary = 0

    def get_data(self):
        self.get_input_data()   
        self.salary = int(input("Enter salary: "))

    def display_details(self):
        print("\n--- Employee Details ---")
        print("Name:", self.name)
        print("Salary:", self.salary)



emp = Employee()

emp.get_data()
emp.display_details()
```
